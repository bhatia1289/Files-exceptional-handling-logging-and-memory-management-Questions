```python
#ans-21
word_to_find = input("Enter the word to search: ")

count = 0

with open("my_file.txt", "r") as f:
    for line in f:
        words = line.split()
        for word in words:
            if word == word_to_find:
                count += 1

print(f"The word '{word_to_find}' occurred {count} times.")
```
